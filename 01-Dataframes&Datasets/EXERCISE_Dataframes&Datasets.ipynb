```python
movie_titles = pd.read_csv("data/movie_titles.tsv")
```

```python
# movie_titles data frame'ini yazdırdığımızda seperator olarak \t kullanıldığını göreceğiz.
movie_titles
```

```python
# Data frame'i düzenlemek için sep paremetresini kullanalım.
movie_titles = pd.read_csv("data/movie_titles.tsv", sep="\t")
```

```python
movie_titles
```

```python
# Yukarıda görüldüğü üzere ilk veri satırı, sütun ismi olarak kullanıldı.
# Yeni sütun isimleri kullanarak bu durumu düzeltelim.
labels = ["ID", "TITLE", "YEAR", "IMDB_RATING", "IMDB_ID", "GENRES"]
movie_titles = pd.read_csv("data/movie_titles.tsv", sep="\t", names=labels)
```
