## $between()$ $Metodunun$ $Filtreleme$ $İşlemlerinde$ $Kullanımı$
***

**`Series.between(left, right, inclusive='both')`, metodun prototipi soldaki gibidir. *between()* metodu, hedeflenen seri için, belirtilen aralıklarda değer bulundurup bulundurmadığına göre True ya da False değerlerinden oluşan bir bool vektör döner. `inclusive` parametresi, "left", "right", "both" ve "neither" değişkenlerini alabilir. Default olarak "both" değişkeni atanmıştır. "inclusive" parametresine "left" verilirse [left, right), "right" verilirse (left, right], "both" verilirse [left, right], "neither" verilirse (left, right) mantığı ile filtreleme yapılır.**

```python
houses.head()
```

```python
# 4 dahil edilmez
houses["bedrooms"].head().between(2, 4, inclusive="left")
```

```python
# 2 dahil edilmez
houses["bedrooms"].head().between(2, 4, inclusive="right")
```

```python
# 2 ve 4 dahil edilmez
houses["bedrooms"].head().between(2, 4, inclusive="neither")
```

```python
# 2 ve 4 dahil edilir
houses["bedrooms"].head().between(2, 4, inclusive="both")
```

```python
# 5 ve 7 adet aralığında yatak odasına sahip olan evlerin bool vektörünü görelim.
houses["bedrooms"].between(5, 7)
```

```python
# 5-7 adet aralığında yatak odasına sahip olan ve olmayan ev sayılarını ayrı ayrı görelim.
houses["bedrooms"].between(5, 7).value_counts()
```

```python
# 5 ve 7 adet aralığında yatak odasına sahip olan evleri görelim.
houses[houses["bedrooms"].between(5, 7)]
```

```python
# 5 ve 7 adet aralığında yatak odasına sahip olan evlerin fiyatlarını görelim.
houses[houses["bedrooms"].between(5, 7)]["price"]
```

```python
# 5 ve 7 adet aralığında yatak odasına sahip olan evlerin fiyatlarını azalan sırada görelim.
houses[houses["bedrooms"].between(5, 7)]["price"].sort_values(ascending=False)
```

```python
houses[houses["grade"].between(11,13)]
```

## $isin()$ $Metodunun$ $Filtreleme$ $İşlemlerinde$ $Kullanımı$
***

**Metodun prototipi: `Series.isin(values)`**

**Kendisine verilen liste içerisinde bulunan değerlerin, üzerinde çalıştığı seri içerisinde olup olmadığına bakar ve buna göre geriye bool vektörü döner. Metoda parametre olarak string tipinde bir değişken verilirse hata fırlatır.**

```python
# Bir ülke listesi oluşturalım.
countries = ["India", "Japan", "South Korea"]
```

```python
# Oluşturduğumuz ülke listesi içerisindeki değerleri "country" sütununu baz alarak DataFrame satırları ile karşılaştıralım.
# Buna göre geri dönen bool vektörünü görüntüleyelim.
netflix["country"].isin(countries)
```

```python
# Bool vektörünü "netflix" DataFrame'ine atayalım.
# Buna göre True değerine sahip satırları görüntüleyelim.
netflix[netflix["country"].isin(countries)]
```

```python
# Belirttiğimiz ülkeler için True değerine sahip satırları, video içerik başlığı, video içerik tipi ve çekimin yapıldığı ülke bazında inceleyelim.
netflix[netflix["country"].isin(countries)][["title", "type", "country"]]
```

```python
mature = netflix["rating"].isin(["TV-MA", "R", "PG-13"])
```

```python
netflix[mature]
```

## $Koşul$ $İfadelerini$ $\&$ $Mantıksal$ $Operatörü$ $ile$ $Filtreleme$ $İşlemi$
***

```python
women = titanic["sex"] == 'female'
died = titanic["survived"] == 0
```

```python
# Titanic kazasında ölen kadınları listeleyelim.
titanic[women & died]
```

```python
# Lebiderya olup fiyatı 500.000'den az olan evleri listeleyelim.
houses[(houses["waterfront"] == 1) & (houses["price"] < 500_000)]
```

```python
houseView = houses["view"] == 4
houseGrade = houses["grade"] >= 11
```

```python
houses[(houseGrade) & (houseView)]
```

```python
high_quality = houses["grade"] >= 11
good_view = houses["view"] == 4
smaller = houses["sqft_living"] <= 3000
```

```python
houses[high_quality & good_view & smaller]
```

```python
houses[houses["yr_built"] >= 2014]
houses[houses["yr_renovated"] >= 2014]
houses[(houses["yr_built"] >= 2014) | (houses["yr_renovated"] >= 2014)]
```

```python
netflix[netflix["director"] == "David Fincher"]
netflix[netflix["director"] == "Martin Scorsese"]
netflix[(netflix["director"] == "David Fincher") | (netflix["director"] == "Martin Scorsese")]
netflix[netflix["director"].isin(["David Fincher", "Martin Scorsese"])]
```

```python
fincher = netflix["director"] == "David Fincher"
scorsese = netflix["director"] == "Martin Scorsese"
recent = netflix["release_year"] > 2015
netflix[(fincher | scorsese) & recent]
```

```python
df = titanic.head()
women = df.sex == 'female'
df[~women]
```

```python
newly_built = houses["yr_built"] >= 2014
newly_renovated = houses["yr_renovated"] >= 2014
recent_homes = newly_built | newly_renovated
houses[recent_homes]
houses[~recent_homes]
```

```python
titanic[titanic.survived == 0]
titanic[~(titanic.survived == 0)]
```
