**pandas.DataFrame.set_index** metodu, veri çerçevesi içerisindeki mevcut sütunlardan birini *index* olacak şekilde kullanabilmemizi sağlar. <a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.set_index.html#pandas-dataframe-set-index">Daha detaylı bilgi için tıklayın.</a>

```python
# Üzerinde çalışacağımız "coin_Bitcoin.csv" veri setini "btc" değişkenine import edelim.
btc = pd.read_csv("../Sources/data/coin_Bitcoin.csv")
```

```python
btc
```

```python
# "btc" veri çerçevesinin index aralığını görelim.
btc.index
```

**`btc` DataFrame'i üzerinde çeşitli incelemeler yaptıktan sonra default olarak atanan index sütunu yerine `date` sütununu yerleştirelim.**

```python
btc.set_index("Date")
```

**Gerçekleştirilen işlem sonucunda elde edilen yeni DataFrame'i yukarıda görebilirsiniz. Ancak bu DataFrame, orijinal `btc` DataFrame'inin bir kopyasıdır. Bu durumu aşağıdaki kod hücresinde yapılan işlem ile doğrulayabiliriz.**

```python
btc
```

**Peki yeni index atamasını orijinal DataFrame üzerinde yapmak için hangi yolu izleyebiliriz? Bunun için *set_index()* metodunun *inplace* parametresini *True* olacak şekilde belirlememiz gerekiyor.**

```python
btc.set_index("Date", inplace=True)
```

```python
btc
```

```python
# "btc" DataFrame'inin ueni index aralığına bakalım.
btc.index
```

```python
btc["High"]
```

```python
# "btc" DataFrame'ini görselleştirerek inceleyelim.
btc["High"].plot(rot=45)
```

## $Sıralama$
***

**Sıralama işlemi yaparken iki adet metot kullanırız. Bunlardan ilki *sort_values()* ve ikincisi *sort_index()*'tir. İlk olarak *sort_values()* metodunu inceleyeceğiz.**

<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html">*sort_values()* metodu hakkında detaylı bilgi için tıklayın.</a>
<a href="https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_index.html">*sort_index()* metodu hakkında detaylı bilgi için tıklayın.</a>

```python
# Üzerinde sıralama işlemi yapacağımız veri setini görelim.
countries
```

```python
# Sıralama işlemini "Healthy life expectancy" sütununa göre yapalım.
# Bu sütun altındaki satırları sıralayacağız.
# sort_values() metodu default olarak azalan sırada sıralama yapar.
countries.sort_values("Healthy life expectancy")
```

```python
# Sıralama işlemimizi artan sırada olacak şekilde yeniden düzenleyelim.
countries.sort_values("Healthy life expectancy", ascending=False)
```

```python
# Üzerinde çalıştığımız veri setini yeniden inceleyelim.
# Bakalım yaptığımız işlemler kalıcı mıydı?
countries
```

```python
# Görüldüğü üzere veri setimiz yaptığımız işlemlerden etkilenmedi.
# Bunu gerçekleştirebilmek için "inplace" parametresine "True" değerini vermeliyiz.
countries.sort_values("Healthy life expectancy", ascending=False, inplace=True)
```

```python
countries
```

```python
# Üzerinde çalışacağımız veri setlerini import edelim.
houses = pd.read_csv("../Sources/data/kc_house_data.csv")
titanic = pd.read_csv("../Sources/data/titanic.csv")
```

```python
# Ev fiyatlarını azalan sırada sıralayalım.
houses.sort_values("price", ascending=False)
```

```python
# Evlerin yatak odası sayılarını azalan sırada sıralayalım.
houses.sort_values("bedrooms", ascending=False)
```

```python
# Şimdi de iki farklı sütunu aynı anda gösterecek şekilde bir sıralama yapalım.
# Evlerin yatak odası sayısını azalan sırada sıralayacağız.
# Bunun yanında her bir yatak odası için banyo sayısını da göstereceği
houses.sort_values(["bedrooms", "bathrooms"], ascending=False)
```

```python
# Şimdi de evlerin banyo sayısını azalan sırada sıralayacağız.
# Bunun yanında her bir banyo için yatak odası sayısını da göstereceğiz.
houses.sort_values(["bathrooms", "bedrooms"], ascending=False)
```

```python
# Titanic veri setini inceleyelim.
titanic
```

```python
# Titanic veri setini isme göre sıralamayı deneyelim.
# Herhangi bir hata almadan alfabetik olarak sıralayacaktır.
# Ancak ufak bir sorunla karşılaşacağız. Küçük harfle başlayan isimler en sona atılacak.
titanic.sort_values("name")
```

```python
# Bunu engellemek için "key" parametresini kullanacağız.
# key parametresi içerisine bir lambda fonksiyonu yazdık, bu fonksiyon bir sütun alıp bu sütun içerisindeki string verileri küçük harfe çeviriyor.
# Bu işlemlerin ardından sıralama gerçekleştiriliyor.
# Sonucu incelediğinizde sorunsuz bir şekilde sıralama yapıldığını görebilirsiniz.
titanic.sort_values("name", key = lambda col: col.str.lower())
```

***
**sort_index() metodunu inceleyelim.**

```python
# Countries veri seti üzerinde işlem yapacağız.
countries
```

```python
# Ülkeler alfabetik sırada olduğu için sort_index() metodunun farkını göremeyeceğiz. Bu nedenle ters sıralama yapalım ve yeniden inceleyelim.
countries.sort_values("Country name", ascending=False, inplace=True)
```

```python
countries
```

```python
# Şimdi de sort_index() metodunu inceleyelim.
countries.sort_index()
```

```python
countries.sort_index(ascending=False, inplace=True)
```

```python
countries
```

```python
titanic.sort_index(ascending=False)
```

**`sort_values()` metodu, sütun değerlerine göre sıralama yaparken, `sort_index()` metodu ise DataFrame'in index değerlerine göre sıralama yapar. Aşağıdaki iki örneği incelediğinizde farkı daha iyi anlayacaksınız.**

```python
titanic["pclass"].value_counts().sort_values()
```

```python
titanic["pclass"].value_counts().sort_index()
```

## $Sıralama$$ve$$Görselleştirme$
***

**Görüldüğü üzere `value_counts()` metodu sonucunda azalan sırada olacak şekilde sıralanmış bir veri elde ediyoruz. Bu durumda index değerleri karışıyor. Veriyi bu şekilde görselleştirdiğimizde yine aynı karışıklığı görebiliriz.**

```python
titanic["pclass"].value_counts()
```

```python
titanic["pclass"].value_counts().plot(kind="bar")
```

**`sort_values()` metodu ile sıralama yaptığımızda da aynı sonuçla karşılaşıyoruz.**

```python
titanic["pclass"].value_counts().sort_values()
```

```python
titanic["pclass"].value_counts().sort_values().plot(kind="bar")
```

**Bunun önüne geçmek için `sort_index()` metodu kullanışlıdır.**

```python
titanic["pclass"].value_counts().sort_index()
```

```python
titanic["pclass"].value_counts().sort_index().plot(kind="bar")
```

```python
houses["bedrooms"].value_counts().sort_values().plot(kind="bar")
```

```python
houses["bedrooms"].value_counts().sort_index().plot(kind="bar")
```

# $iloc[]$ $Metodu$
***


**`iloc[]` metodu, tıpkı `loc[]` metodunda olduğu gibi, satırların isimlerine göre, o satırın sütun değerlerini görmemizi sağlar. Ancak bunu yaparken string değerleri yerine tam sayı değerleri parametre olarak alır. Bu tam sayı değerleri, DataFrame'in index'leridir.**

```python
countries
```

```python
countries.iloc[0]
```

```python
type(countries.iloc[0])
```

```python
countries.iloc[[0]]
```

```python
type(countries.iloc[[0]])
```

```python
countries.iloc[[10,20,30,100]]
```

```python
countries.iloc[50:60]
```

```python
houses.sort_index(ascending=False, inplace=True)
```

```python
houses
```

**Aşağıdaki iki örnek, loc ve iloc arasındaki farkı açıklamaktadır. loc metodu, DataFrame'in etiket isimlerini kullanarak işlem yaparken, iloc metodu DataFrame'in etiket indexlerini kullanarak işlem yapar.**

```python
houses.iloc[0:5]
```

```python
houses.loc[21612:21611]
```
